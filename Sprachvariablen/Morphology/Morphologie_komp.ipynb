(1sg, 2sg, 3sg, 1pl, 2pl, 3pl, n3)

```python
import stanza
import string

stanza.download('la') # Download the Latin model

nlp = stanza.Pipeline(lang='la', package='proiel', processors='tokenize,mwt,pos', verbose=False) # Load the Latin model

def verb_ann(sentence):
    
    # Remove punctuation marks from sentence
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))
    
    # to lower case
    sentence = sentence.lower()

    # Split sentence into words
    words = sentence.split()
    
    doc = nlp(sentence)
    annotations = []
    person_1sg = []
    person_2sg = []
    person_3sg = []
    person_1pl = []
    person_2pl = []
    person_3pl = []
    
    for sent in doc.sentences:
        for word in sent.words:
            feats = word.feats
            if feats is not None and 'VerbForm=Fin' in feats:
                annotations.append(feats)
                if 'Person' in feats and 'Number' in feats:
                    feats_dict = {key: value for key, value in [f.split('=') for f in feats.split('|')]}
                    if feats_dict['Person'] == '1' and feats_dict['Number'] == 'Sing':
                        person_1sg.append(word.text)
                    elif feats_dict['Person'] == '2' and feats_dict['Number'] == 'Sing':
                        person_2sg.append(word.text)
                    elif feats_dict['Person'] == '3' and feats_dict['Number'] == 'Sing':
                        person_3sg.append(word.text)
                    elif feats_dict['Person'] == '1' and feats_dict['Number'] == 'Plur':
                        person_1pl.append(word.text)
                    elif feats_dict['Person'] == '2' and feats_dict['Number'] == 'Plur':
                        person_2pl.append(word.text)
                    elif feats_dict['Person'] == '3' and feats_dict['Number'] == 'Plur':
                        person_3pl.append(word.text)

                    
                    
    verbcount = len(annotations)
    
    person_1sg_count = len(person_1sg)
    person_2sg_count = len(person_2sg)
    person_3sg_count = len(person_3sg)
    person_1pl_count = len(person_1pl)
    person_2pl_count = len(person_2pl)
    person_3pl_count = len(person_3pl)
    
    person_not_3_count = len(annotations) - len(person_3sg) - len(person_3pl)
    
    ratio_1sg = round(person_1sg_count / verbcount,3)
    ratio_2sg = round(person_2sg_count / verbcount,3)
    ratio_3sg = round(person_3sg_count / verbcount,3)
    ratio_1pl = round(person_1pl_count / verbcount,3)
    ratio_2pl = round(person_2pl_count / verbcount,3)
    ratio_3pl = round(person_3pl_count / verbcount,3)
    
    ratio_not_3 = round(person_not_3_count / verbcount,3)
    
    
    return ratio_1sg, ratio_2sg, ratio_3sg, ratio_1pl, ratio_2pl, ratio_3pl, ratio_not_3
```

(nom, gen, dat, acc, abl)

```python
import stanza
import string

stanza.download('la') # Download the Latin model

nlp = stanza.Pipeline(lang='la', package='proiel', processors='tokenize,mwt,pos', verbose=False) # Load the Latin model

def noun_ann(sentence):
    
    # Remove punctuation marks from sentence
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))
    
    # to lower case
    # sentence = sentence.lower()

    # Split sentence into words
    words = sentence.split()
    
    doc = nlp(sentence)
    annotations = []
    nom = []
    gen = []
    dat = []
    acc = []
    abl = []
    
    for sent in doc.sentences:
        for word in sent.words:
            upos = word.upos
            feats = word.feats
            if upos is not None and 'NOUN' in upos:
                annotations.append(upos)
                if 'Case' in feats:
                    feats_dict = {key: value for key, value in [f.split('=') for f in feats.split('|')]}
                    if feats_dict['Case'] == 'Nom':
                        nom.append(word.text)
                    elif feats_dict['Case'] == 'Gen':
                        gen.append(word.text)
                    elif feats_dict['Case'] == 'Dat':
                        dat.append(word.text)
                    elif feats_dict['Case'] == 'Acc':
                        acc.append(word.text)
                    elif feats_dict['Case'] == 'Abl':
                        abl.append(word.text)
                    
    nouncount = len(annotations)
    
    nom_count = len(nom)
    gen_count = len(gen)
    dat_count = len(dat)
    acc_count = len(acc)
    abl_count = len(abl)
    
    ratio_nom = round(nom_count / nouncount,3)
    ratio_gen = round(gen_count / nouncount,3)
    ratio_dat = round(dat_count / nouncount,3)
    ratio_acc = round(acc_count / nouncount,3)
    ratio_abl = round(abl_count / nouncount,3)
    
    
    return ratio_nom, ratio_gen, ratio_dat, ratio_acc, ratio_abl
```
